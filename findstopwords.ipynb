```python
tp=open('senttemp1.txt', 'r',encoding="utf-8")
s=tp.readlines()
x={}
for i in range(len(s)):
    t1=s[i].split(' ')
    for i in t1:
        if i not in x:
            x[i]=0
        x[i]+=1
tp.close()
x
```

```python
import pandas as pd
df =pd.DataFrame(x.items())
```
